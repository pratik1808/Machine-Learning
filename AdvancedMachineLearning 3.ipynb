# Question 3: PCA (8 pts)

Explain the principle of Principal Component Analysis algorithm and in what sense is the selection of the eigenvectors to represent the data is an optimal choice.  How do you reconstruct (a noisy version of) the original data from the eigenvectors and the loadings?



#### Answer:

Principal component analysis (PCA) is a statistical procedure that is used to reduce the dimensionality. The steps taken by PCA to do the same are as follows:
1. PCA computes eigenvectors of the covariance matrix, which can be referred as "principal axes"
2. It the sorts the "principal axes" by their eigenvalues. Eigenvalues is basically the amount of explained variance. 
3. The centered data can then be projected onto these principal axes to yield principal components. The prinicpal components are the "scores". 
4. For the purposes of dimensionality reduction, one can keep only a subset of principal components and discard the rest. 

The above steps can be explained algorithmically below:

1. Let ùêó1 be the ùëõ√óùëù data matrix with ùëõ rows, each row is a data point and ùëù columns, where each column represents a variable or a feature. 
2. After subtracting the mean vector ùùÅ from each row, we get the centered data matrix ùêó1. 
3. Let ùêï be the ùëù√óùëò matrix of some ùëò eigenvectors that we want to use; these would most often be the ùëò eigenvectors with the largest eigenvalues. The selection of only the vectors we would like to use results ultimately in dimension reduction
4. Then the ùëõ√óùëò matrix of PCA projections will be simply given by ùêô=ùêó1ùêï. These PCA projections are the "scores"

The reason we use eigenvectors is to calculate distances between data points. This can be useful when creating models that need to group data points based on similarity. Eigenvalues can hence be used as weights in PCA and can be the optimal choice.

In order to be able to reconstruct the original data from principal components, lets consider the above algorithm's output and follow the steps:
1. We can map the ùêô matrix back to ùëù dimensions with transpose of ùêï. 
2. The values of each principal component should be placed on the same vector as was used for projection

````python
np.random.seed(18)
sample_set = np.random.choice(a= data['SalePrice'], size=800)
#Sample mean
sample_mean = np.mean(sample_set)
#Population mean
population_mean = np.mean(data_dm['SalePrice'])

print("Sample Mean: ", str(sample_mean)+"\nPopulation Mean: "+str(population_mean))
````

e) i)**(2pts)** Calculate 95% confidence intervals for SalePrice with a sample size of 100. 

ii)**(3pts)** Calculate 95 % confidence intervals for 100 different trials with a sample size of 500. Plot the confidence intervals and interpret how it captures the population mean.

````python
np.random.seed(10)
sample_size = 100
sample = np.random.choice(a= data_dm['SalePrice'],
                          size = sample_size)
sample_mean = sample.mean()

#Get the critical Z value
z_critical = 1.96 # for 95 % confidence interval

#Get population standard deviation
pop_stdev = np.std(data['SalePrice'])

#margin of error
margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size)) 

#confidence interval
confidence_interval = (sample_mean - margin_of_error,
                       sample_mean + margin_of_error)  

#Print confidence interval and true mean value
print("Confidence Interval :",confidence_interval)
print("Population Mean :",population_mean)

````

````python
np.random.seed(12)

#define the sample size
sample_size=500
intervals = []
sample_means = []
z_critical = 1.96 #intializing outsize loop as it's fixed for a 95% interval
pop_stdev = np.std(data['SalePrice']) #calculating outside the loop as SD of population does not change with sample

for sample in range(100):
    sample = np.random.choice(a= data_dm['SalePrice'], size = sample_size)
    sample_mean = sample.mean()
    sample_means.append(sample_mean)
    margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size))
    confidence_interval = (sample_mean - margin_of_error,
                           sample_mean + margin_of_error)  
    intervals.append(confidence_interval)
    
plt.figure(figsize=(13, 9))

plt.errorbar(x=np.arange(0.1, 100, 1), 
             y=sample_means, 
             yerr=[(top-bot)/2 for top,bot in intervals],
             fmt='o')

plt.hlines(xmin=0, xmax=100,
           y=data['SalePrice'].mean(), 
           linewidth=2.0,
           color="red")
plt.title('Confidence Intervals for 100 Trials', fontsize = 20)
plt.show()
````

As we can see above, sample mean of the samples taken are approximately around the value of the population mean(redline). This verifies the Central Limit Theorem and sampling theory as well. This suggests that in real world datasets, where we do not have actual whole population data, we can use the sample mean and other metrics and specify the confidence interval to arrive at insightful results.
